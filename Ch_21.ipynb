# How to detect the Specific color in Python

```python
# Import the library
import cv2 as cv
import numpy as np

# Define a callback function for trackbar, which accepts an argument
def slider(x):
    pass

# Path to the image
path = "resources/room.jpg"

# Create a window for the sliders
cv.namedWindow("Bars")
cv.resizeWindow("Bars", 720, 300)

# Creating trackbars in the "Bars" window
cv.createTrackbar("Hue Min", "Bars", 0, 179, slider)
cv.createTrackbar("Hue Max", "Bars", 179, 179, slider)
cv.createTrackbar("Sit Min", "Bars", 0, 255, slider)
cv.createTrackbar("Sit Max", "Bars", 255, 255, slider)
cv.createTrackbar("Val Min", "Bars", 0, 255, slider)
cv.createTrackbar("Val Max", "Bars", 255, 255, slider)

while True:
    # Read the image
    img = cv.imread(path)

    # Check if the image is loaded properly
    if img is None:
        print(f"Error: Could not read image from path: {path}")
        break

    # Convert the image to HSV
    hsv_img = cv.cvtColor(img, cv.COLOR_BGR2HSV)
    
    # Get the current positions of the trackbars
    hue_min = cv.getTrackbarPos("Hue Min", "Bars")
    hue_max = cv.getTrackbarPos("Hue Max", "Bars")
    sit_min = cv.getTrackbarPos("Sit Min", "Bars")
    sit_max = cv.getTrackbarPos("Sit Max", "Bars")
    val_min = cv.getTrackbarPos("Val Min", "Bars")
    val_max = cv.getTrackbarPos("Val Max", "Bars")
    
    # Print the HSV values (for debugging)
    print(hue_min, hue_max, sit_min, sit_max, val_min, val_max)
    
    # Define lower and upper HSV range based on trackbar positions
    lower = np.array([hue_min, sit_min, val_min])
    upper = np.array([hue_max, sit_max, val_max])
    
    # Create a mask using the lower and upper HSV values
    mask = cv.inRange(hsv_img, lower, upper)
    result = cv.bitwise_and(img, img, mask=mask)
    
    # Show the original, HSV, mask, and result images
    cv.imshow("Original", img)
    cv.imshow("HSV", hsv_img)
    cv.imshow("Mask", mask)
    cv.imshow("Result", result)
    
    # Break the loop if 'q' is pressed
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

# Clean up windows
cv.destroyAllWindows()
```
